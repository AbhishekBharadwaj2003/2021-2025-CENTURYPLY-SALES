```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_excel("Sales Data-Jan'21 to Dec'21.xlsx", sheet_name="Sheet1")
```

```python
df.drop(columns=[ 'PO.No', 'PI.No', 'Brand', 'Sub-Brand', 'Prime Brand', 'Folder/Non-Folder', 'Design(LAM)', 'Finish(LAM)'], inplace=True)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Month'] = df['Date'].dt.to_period('M').astype(str)
df['Weekday'] = df['Date'].dt.day_name()
df['Day'] = df['Date'].dt.day
df['Week'] = df['Date'].dt.isocalendar().week
df['Quarter'] = df['Date'].dt.quarter
```

```python
monthly_sales = df.groupby('Month')['Billed Qty(PCS)'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly_sales, x='Month', y='Billed Qty(PCS)', marker='o', color='green')
plt.title('Monthly Sales Trend (PCS)', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Total PCS')
plt.tight_layout()
plt.show()
```

```python
weekly_sales = df.groupby('Week')['Billed Qty(PCS)'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=weekly_sales, x='Week', y='Billed Qty(PCS)', marker='o', color='orange')
plt.title('Weekly Sales Trend (PCS)', fontsize=14)
plt.xlabel('Week Number')
plt.ylabel('Total PCS')
plt.tight_layout()
plt.show()
```

```python
df['Weekday'] = df['Date'].dt.day_name()
```

```python
weekday_sales = df.groupby('Weekday')['Billed Qty(PCS)'].sum().reindex([ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
]).reset_index()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=weekday_sales, x='Weekday', y='Billed Qty(PCS)', palette='Blues_r')
plt.title('Sales by Day of the Week (PCS)', fontsize=14)
plt.xlabel('Day')
plt.ylabel('Total PCS')
plt.tight_layout()
plt.show()
```

```python
product_sales = df.groupby('Material Desc.')['Billed Qty(PCS)'].sum().reset_index()
```

```python
product_sales_sorted = product_sales.sort_values(by='Billed Qty(PCS)', ascending=False)
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=product_sales_sorted.head(15), x='Billed Qty(PCS)', y='Material Desc.', palette='Blues')
plt.title('Top 15 Products by Sales Volume (PCS)', fontsize=14)
plt.xlabel('Total Billed Qty (PCS)')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=product_sales_sorted.tail(15), x='Billed Qty(PCS)', y='Material Desc.', palette='rocket')
plt.title('Bottom 15 Products by Sales Volume (PCS)', fontsize=14)
plt.xlabel('Total Billed Qty (PCS)')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
territory_sales = df.groupby('Territory')['Billed Qty(PCS)'].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=territory_sales.head(15), x='Billed Qty(PCS)', y='Territory', palette='mako')
plt.title('Top 15 Territories by Sales Volume (PCS)', fontsize=14)
plt.xlabel('Total PCS Sold')
plt.ylabel('Territory')
plt.tight_layout()
plt.show()
```

```python

top_territories = territory_sales.head(8).set_index('Territory')
top_territories['Billed Qty(PCS)'].plot.pie(autopct='%1.1f%%', figsize=(8, 8), startangle=140)
plt.title('Top 8 Territories by Sales Share (PCS)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python

unique_customers = df['Customer Name'].nunique()
print("🔢 Number of Unique Customers:", unique_customers)
```

```python
top_customers = df.groupby('Customer Name')['Billed Qty(PCS)'].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_customers.head(15), x='Billed Qty(PCS)', y='Customer Name', palette='flare')
plt.title('Top 15 Customers by Sales Volume (PCS)', fontsize=14)
plt.xlabel('Total PCS Sold')
plt.ylabel('Customer Name')
plt.tight_layout()
plt.show()
```

```python
customer_freq = df['Customer Name'].value_counts().reset_index()
customer_freq.columns = ['Customer Name', 'Number of Transactions']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=customer_freq.head(15), x='Number of Transactions', y='Customer Name', palette='magma')
plt.title('Top 15 Customers by Frequency of Transactions', fontsize=14)
plt.xlabel('Number of Invoices')
plt.ylabel('Customer Name')
plt.tight_layout()
plt.show()
```
