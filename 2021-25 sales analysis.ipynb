```python
# importing the necessary packages

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Reading Multi-Year Sales Data from Excel Files Using Pandas

df1 = pd.read_excel("Sales Data-Jan'21 to Dec'21.xlsx", sheet_name="Sheet1")
df2 = pd.read_excel("Sales Data - Jan'22 to Mar'23.xlsx", sheet_name="Sheet1")
df3 = pd.read_excel("Sales Data-Apr'23 to Jan'25.xlsx", sheet_name="Sheet1")
```

```python
# Merging Multiple Excel Files into a Single DataFrame

df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
#Converting 'Date' Column to DateTime Format Using Pandas

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Ensuring Consistent Naming for Quantity Column

df.rename(columns={"Billing Qty": "Billing Qty (pcs)"}, inplace=True)
```

```python
# top 5 rows

print(df.head())
```

```python
# Checking for Missing Values in the Dataset

print(df.isnull().sum())
```

```python
# show the column names

print(df.columns)
```

```python
# Standardizing Column Names for Consistency

df.rename(columns={"Billing Qty": "Billing Qty (pcs)"}, inplace=True)
```

```python
# Removing Rows with Missing Date or Quantity Values

df.dropna(subset=["Date", "Billed Qty"], inplace=True)
```

```python
# Displaying DataFrame Structure and Data Types

df.info()
```

```python
# Generating Summary Statistics for Numerical Columns

df.describe()
```

```python
# Calculating Monthly Sales Quantity Using GroupBy

monthly_sales = df.groupby(df['Date'].dt.to_period('M'))["Billed Qty(PCS)"].sum()
```

```python
# Visualizing Monthly Sales Trend Using Line Plot

monthly_sales.plot(kind='line', figsize=(15, 5), marker='o', title="Monthly Sales Trend")
plt.ylabel("Billed Qty(PCS)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Identifying Top 10 Customers by Sales Volume

top_customers = df.groupby("Customer Name")["Billed Qty(PCS)"].sum().nlargest(10)
```

```python
# Bar Chart of Top 10 Customers by Quantity Sold

top_customers.plot(kind='bar', title='Top 10 Customers by Billing Qty (pcs)', figsize=(10, 5))
plt.ylabel("Billing Qty (pcs)")
plt.xticks(rotation=45)
plt.show()
```

```python
#  Calculating Total Sales Quantity by Territory

territory_sales = df.groupby("Territory")["Billed Qty(PCS)"].sum()
```

```python
# Visualizing Sales Distribution Across Territories

    territory_sales.plot(kind='bar', title='Sales by Territory', figsize=(35, 15))
    plt.ylabel("Billing Qty (pcs)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Finding Top 10 Best-Selling Products by Quantity

top_products = df.groupby("Material Desc.")["Billed Qty(PCS)"].sum().nlargest(10)
```

```python
#  Printing All Column Names as a List

print(df.columns.tolist())
```

```python
#  Bar Chart of Top 10 Products by Quantity Sold

top_products.plot(kind='bar', title='Top 10 Products by Sales (pcs)', figsize=(10, 5), color='orange')
plt.ylabel("Billing Qty (pcs)")
plt.xlabel("Product Name")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculating the Average Quantity Billed per Order

avg_billing_qty = df["Billed Qty(PCS)"].mean()
print(f" Average Billing Qty per Order: {avg_billing_qty:.2f} pcs")
```

```python
#  Identifying Top 10 Territories by Sales Volume

top_Territories = df.groupby("Territory")["Billed Qty(PCS)"].sum().nlargest(10)
```

```python
#  Bar Chart of Top 10 Territories by Quantity Sold

top_Territories.plot(kind='bar', title='Top 10 Cities by Sales (pcs)', figsize=(10, 5), color='teal')
plt.ylabel("Billing Qty (pcs)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Finding Top 10 Customers by Number of Orders

customer_orders = df["Customer Name"].value_counts().nlargest(10)
```

```python
#Bar Chart of Top 10 Most Frequent Customers

customer_orders.plot(kind='bar', title='Top 10 Customers by Order Frequency', figsize=(10, 5), color='purple')
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
